```python
import numpy as np
from matplotlib import pyplot as plt
def generation_sin(f0,fe,N):
    t=np.linspace(0,N/fe,N) 
    signal=np.sin(2*np.pi*f0*t)
    plt.plot(t, signal)
    plt.title(f"Signal sinusodal : f0 = {f0} Hz, fe = {fe} Hz, N = {N}")
    plt.xlabel("Temps (s)")
    plt.ylabel("Amplitude")
    plt.show()
    return(signal)
f0=1
fe=100
N=500
signal=generation_sin(f0, fe, N)
```

```python
#Q2
def energie(x):
    xt=np.transpose(x)
    ene=x.dot(xt)
    return(ene)
print(energie(signal))
def puissance(x):
    #la puissance est l'nergie / temps
    N=len(x)
    xt=np.transpose(x)
    puissance=2*x.dot(xt)/(2*N+1)
    return(puissance)
print("la puissance estime par le calcul " +str(puissance(signal)))
```

Commentaire: la puissance thorique moyenne d'un signal sinusoidal = 0.5
             

```python
def quant(n,N,signal,fe,f0):
    niveau=2**n
    x_max=np.max[signal]
    x_min=np.min[signal]
    pas=(x_max - x_min) / niveau
    signal2=np.round((signal-x_min)/pas) * pas + x_min
    t=np.linspace(0,N/fe,N)
    plt.plot(t,signal2)
    plt.show()
    return(signal2)
quant(3,N,signal,fe,f0)
```
